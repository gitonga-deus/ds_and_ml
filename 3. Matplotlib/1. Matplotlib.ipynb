```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x, y)
```

# Subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
#Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes_1.plot(x,y)
axes_2.plot(y,x)

axes_2.set_title("SMALLER PLOT")
axes_1.set_title("LARGER PLOT")
```

```python
# Creating Subplots using Object Oriented Method
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_title("First Plot")

axes[1].plot(y,x,'b')
axes[1].set_title("Second Plot")

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'g')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(7,4), nrows=2, ncols=1)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("my_plot.png")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared', linewidth=1, ls="--", marker='o')
ax.plot(x,x**3, label='X Cubed',linewidth=1, ls="--", marker='o')

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')

ax.legend(loc=0)
```
