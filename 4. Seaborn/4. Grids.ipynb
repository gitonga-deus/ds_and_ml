```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
my_plot = sns.PairGrid(iris)

my_plot.map_diag(sns.distplot)
my_plot.map_upper(plt.scatter)
my_plot.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
my_plot_1 = sns.FacetGrid(tips, col='time', row='smoker')
 
my_plot_1.map(sns.distplot, 'total_bill')
```
